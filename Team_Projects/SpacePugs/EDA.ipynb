# Reforms

```python
# # From share option in google sheets
# https://docs.google.com/spreadsheets/d/145OhTzEq9xfLW-nDeyq0cUphbg6MLtGU/edit?usp=sharing&ouid=108255493032319164983&rtpof=true&sd=true -- direct Reforms sheet
# # Tab found in url
sheet_id = "145OhTzEq9xfLW-nDeyq0cUphbg6MLtGU"
tab_id = "890748922"
# pd.set_option('display.max_rows', 10)

reforms = pd.read_csv(f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv&gid={tab_id}")
```

```python
reforms.groupby('Region').size()
```

```python
pd.set_option('display.max_rows', 10)
reforms.groupby(['Type of Reform', 'Economy']).size()
```

```python
reforms_subsaharan = reforms[reforms['Region'] == 'Sub-Saharan Africa']
```

```python
reforms_subsaharan.groupby('Economy').size()
```

```python
reforms_subsaharan.groupby(['Type of Reform', 'Economy']).size()
```

```python
counts = reforms_subsaharan.groupby(['Type of Reform', 'Economy']).size()
counts = counts.reset_index(name='Count')
counts
```

```python
xtab = pd.crosstab(reforms_subsaharan['Economy'], reforms_subsaharan['Type of Reform'])
myplot = xtab.plot(kind='barh', figsize = [15,10], stacked=True)
plt.ylabel('Country')
plt.title('Reform Types by Country in Subsaharan Africa')
```
